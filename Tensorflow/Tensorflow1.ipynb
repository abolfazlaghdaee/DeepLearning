# Neural Netwrok

```python
import tensorflow as tf
from tensorflow import keras
```

### Date Definition



```python
x = tf.constant([-1, -2, 0, 1, 2, 3, 4, 5, 6, 7 ], dtype = 'float32')

y = x * 3 +1
```

```python
x.numpy(), y.numpy()
```

# Model Definition

```python
model = keras.Sequential([
    keras.layers.Dense(units =1 , activation=None, input_shape= [1])
])
```

```python
model.compile(optimizer = 'sgd', loss= 'mean_squared_error')
```

```python
model.summary()
```

```python
model.fit(x,y , batch_size = 1, epochs= 100)
```

```python
x_test = tf.constant([-2, 3, 4])
y_test = x_test * 3 +1
```

```python
y_test.numpy()
```

```python
model.predict(x_test)
```

## Next Session

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

### Data Generation
y = x^2 + x + 3

```python
x = np.arange(-80, 81, 0.5)
y = x **2 +x + 3
```

```python
x.shape , y.shape
```

```python
x_train , x_test, y_train , y_test = train_test_split(x, y , test_size =0.2)
```

```python
x_train.shape, x_test.shape
```

## 1. Frist Model

```python
model_1 = keras.Sequential()
model_1.add(keras.layers.Dense(units= 1024, activation= 'relu', input_shape=[1]))
model_1. add(keras.layers.Dense(units = 1))
```

```python
model_1.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
model_1.summary()
```

```python
hist_1 = model_1.fit(x_train, y_train, epochs = 1000, batch_size = 64, validation_data= (x_test, y_test))
```

## Second Model

```python
model_2 = keras.Sequential()
model_2.add(keras.layers.Dense(units= 256, activation= 'relu', input_shape = [1]))
model_2.add(keras.layers.Dense(units= 256, activation= 'relu'))
model_2.add(keras.layers.Dense(units= 256, activation= 'relu'))
model_2.add(keras.layers.Dense(units= 256, activation= 'relu'))
model_2.add(keras.layers.Dense(units= 1))
```

```python
model_2.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
model_2.summary()
```

```python
model_2.fit(x_train, y_train, batch_size = 64, epochs= 1000, validation_data  = (x_test, y_test))
```

## Comparision

```python
y_model1_pred = model_1.predict(x_test)
y_model2_pred = model_2.predict(x_test)
```

```python
plt.scatter(x_test, y_test, color= 'black', linewidths = 1, marker='*')
plt.scatter(x_test, y_model1_pred, color = 'b')
plt.scatter(x_test, y_model2_pred , color = 'r')
plt.show()
```
